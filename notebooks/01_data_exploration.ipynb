# News Classification - Data Exploration

This notebook explores the FakeNewsCorpus dataset for news classification and fake news detection.

```python
# Import libraries
import sys
import os
sys.path.append('..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
plt.style.use('seaborn-v0_8')
sns.set_palette('husl')
%matplotlib inline
```

```python
# Load data
from src.data.loader import DataLoader

loader = DataLoader()
df = loader.load_fakenews_sample("small")

print(f"Dataset shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
```

```python
# Show first few rows
df.head()
```

```python
# Basic info
df.info()
```

```python
# Check for missing values
missing = df.isnull().sum()
print("Missing values per column:")
print(missing[missing > 0])
```

```python
# Label distribution
label_counts = loader.get_label_distribution(df)
print("Label distribution:")
print(label_counts)
```

```python
# Visualize label distribution
plt.figure(figsize=(12, 6))
label_counts.plot(kind='bar')
plt.title('Distribution of News Types')
plt.xlabel('News Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Text length analysis
df['content_length'] = df['content'].fillna('').str.len()
df['title_length'] = df['title'].fillna('').str.len()

print("Text length statistics:")
print(f"Content - Mean: {df['content_length'].mean():.0f}, "
      f"Std: {df['content_length'].std():.0f}, "
      f"Min: {df['content_length'].min()}, "
      f"Max: {df['content_length'].max()}")

print(f"Title - Mean: {df['title_length'].mean():.0f}, "
      f"Std: {df['title_length'].std():.0f}, "
      f"Min: {df['title_length'].min()}, "
      f"Max: {df['title_length'].max()}")
```

```python
# Visualize text lengths
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Content length distribution
axes[0].hist(df['content_length'], bins=50, edgecolor='black')
axes[0].set_title('Content Length Distribution')
axes[0].set_xlabel('Character Count')
axes[0].set_ylabel('Frequency')

# Title length distribution
axes[1].hist(df['title_length'], bins=30, edgecolor='black')
axes[1].set_title('Title Length Distribution')
axes[1].set_xlabel('Character Count')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Check text length by label
plt.figure(figsize=(12, 6))
sns.boxplot(x='type', y='content_length', data=df)
plt.title('Content Length by News Type')
plt.xlabel('News Type')
plt.ylabel('Content Length (characters)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Domain analysis
domain_counts = df['domain'].value_counts().head(20)
print("Top 20 domains:")
print(domain_counts)
```

```python
# Create train/test split
splits = loader.create_train_test_split(df)

print(f"Train set: {len(splits['train'])} articles")
print(f"Test set: {len(splits['test'])} articles")

# Check label distribution in splits
train_labels = splits['train']['type'].value_counts()
test_labels = splits['test']['type'].value_counts()

print("\nTrain label distribution:")
print(train_labels)
print("\nTest label distribution:")
print(test_labels)
```
